## 2. Extract data from dataset

This is achieved using **netCDF4** package

```python
from netCDF4 import Dataset

file_list = [
    'data/data_Hs_Tp_2015-17',
    'data/data_Hs_Tp_2018-20',
    'data/data_Hs_Tp_2021-23'
]
nc = Dataset(file_list, 'r')
print(nc.variables)
pass
```

```python
lon = nc.variables['longitude']
lat = nc.variables['latitude']
time = nc.variables['valid_time']

for d in nc.dimensions.items():
    print(d)
```

```python
lon_array = lon[:]
lat_array = lat[:]
Time_array = time[:]

print(lon_array)

print(lat_array)

print(Time_array)
pass
```

```python
hs = nc.variables['shww']
tp = nc.variables['pp1d']

hs_array = hs[:]
period_array = tp[:]

print("Hs array")
print(hs_array)

print("Tp array")
print(period_array)
```

```python
i = 0 #for the present case, the values can be 0,1
j = 0 #for the present case, the values can be 0,1
hs_array = hs[:,i,j]
period_array = tp[:,i,j]

print("Hs array")
print(hs_array)

print("Tp array")
print(period_array)
```

#### Exporting to excel sheet

```python
import pandas as pd
import numpy as np


data = np.array([Time_array,
         hs_array,
         period_array])
data = data.T


df = pd.DataFrame(data, columns=['time', 'Hs', 'Tp'])

df.to_csv('hsarrays.csv', index=False)
```
